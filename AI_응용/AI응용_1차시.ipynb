# 사진 또는 영상 필터링

opencv는 기본적으로 색상 채널 배열이 bgr순서로 되어있다.

-------------------------



차원(channel):  이미지.ndim

형태(shape):    이미지.shape

데이터타입:      이미지.dtype

-------------------------

- 크기 변경

cv2.resize(이미지, (가로,세로))

-------------------------

- 색상 전처리

cv2.cvtColor(이미지, 변환 방법)

변환 방법
>cv2.COLOR_BGR2GRAY

컬러 이미지를 흑백(Gray)으로 바꿉니다.

>cv2.COLOR_BGR2RGB

BGR 순서를 RGB 순서로 바꿉니다.

>cv2.COLOR_BGR2HSV

색상(Hue), 채도(Saturation), 명도(Value) 체계로 바꿉니다.

----------------------------------------------------------------

- 블러

>cv2.blur(이미지, (커널 크기))

커널 영역 내의 모든 픽셀 값을 더해 단순 평균을 내는 가장 기본적인 블러 방식입니다.

>cv2.GaussianBlur(이미지,(커널 크기),x축 방향 표준편차,y축 방향 표준편차)

중심부 픽셀에 더 높은 가중치를 두는 가우시안 분포(종 모양)를 사용하여 부드럽게 만듭니다.

>cv2.medianBlur(이미지,커널 크기)

주변 픽셀들을 크기순으로 정렬한 뒤 그중 중간값을 선택하여 현재 픽셀을 바꿉니다.

----------------------------------------------------------------

- 임계값

임계값, 결과이미지 = cv2.threshold(이미지, 임계값, 최대값,임계 방법)

임계 방법
>cv2.THRESH_BINARY

가장 기본,"T보다 크면 최대값(255), 아니면 0

>cv2.THRESH_BINARY_INV

반전 이진화,"T보다 크면 0, 아니면 최대값(255) (흑백 반전)

>cv2.THRESH_TRUNC,자르기

T보다 크면 T, 아니면 원래 값 유지 (밝은 부분이 깎임)

>cv2.THRESH_TOZERO,0으로    

만들기,"T보다 크면 원래 값, 아니면 0 (어두운 부분 제거)

>cv2.THRESH_TOZERO_INV  

반대로 0으로,"T보다 크면 0, 아니면 원래 값 유지"
